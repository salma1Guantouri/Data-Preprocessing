```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Standardisation des données
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Appliquer l'ACP pour différents pourcentages de variance conservée
for variance_percentage in [0.7, 0.9]:
    # Appliquer l'ACP
    pca = PCA(n_components=variance_percentage)
    X_train_pca = pca.fit_transform(X_train_scaled)
    X_test_pca = pca.transform(X_test_scaled)

    # Entraîner le modèle (par exemple, RandomForestClassifier)
    clf = RandomForestClassifier(n_estimators=100, random_state=42)
    clf.fit(X_train_pca, y_train)

    # Prédire et évaluer le modèle
    y_pred = clf.predict(X_test_pca)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Variance conservée : {variance_percentage}, Accuracy : {accuracy}")
```

# **Data déséquilibrées**

**Suite à l'exploration visuelle de notre jeu de données, il est devenu évident que celui-ci présente une distribution déséquilibrée de ses classes.**

**Cette disparité entre les différentes catégories pourrait introduire un biais significatif dans notre modèle. Ainsi, afin de garantir des prédictions précises et équitables, nous allons mettre en œuvre des techniques d'équilibrage des données.**

**Ces méthodes visent à harmoniser la représentation de chaque classe, permettant ainsi au modèle d'apprendre de manière plus équilibrée et de fournir des résultats plus fiables.**

```python
from imblearn.over_sampling import SMOTE
X=df.drop('label', axis=1)
Y=df['label']
# Instanciation de SMOTE
smote = SMOTE(random_state=42)

# Rééchantillonnage du dataset déséquilibré
X_resampled, y_resampled = smote.fit_resample(X, Y)
```

```python
label_counts = y_resampled .value_counts(normalize=True)
plt.figure(figsize=(6, 6))
plt.pie(label_counts, labels=label_counts.index, autopct='%1.1f%%', startangle=140)
plt.title("Répartition des Étiquettes")
plt.axis('equal')  # Assurer que le graphique est un cercle et non une ellipse
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Avant la pondération
print("Avant la pondération:")
print(y_train.value_counts())

# Entraîner le modèle RandomForest avec la pondération des classes
rf_weighted = RandomForestClassifier(class_weight='balanced', random_state=42)
rf_weighted.fit(X_train, y_train)

# Prédire sur l'ensemble de test
y_pred_weighted = rf_weighted.predict(X_test)

# Évaluer les prédictions
print("\nRapport de classification après la pondération:")
print(classification_report(y_test, y_pred_weighted))

# Matrice de confusion
print("\nMatrice de confusion après la pondération:")
print(confusion_matrix(y_test, y_pred_weighted))

```
